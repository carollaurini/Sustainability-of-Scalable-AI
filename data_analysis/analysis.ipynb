```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
```

```python
#n=2
df2 = pd.read_csv("C:/Users/carol/Desktop/RadBoudCode/out_2.txt", sep =';',names = ['Bandwidth','Compute rate (GFLOPS)','Relative error','Runtime'])
df2['MAX_ITER'] = range(1,1001)
df2['N'] = 2
df2
```

```python
# Placing the plots in the plane
plot1 = plt.subplot2grid((6, 6), (0, 0), rowspan=2,colspan=2)
plot2 = plt.subplot2grid((6, 6), (0, 2), rowspan=4,colspan=2)
plot3 = plt.subplot2grid((6, 6), (2, 0), rowspan=2,colspan=2)
plot4 = plt.subplot2grid((6, 6), (0, 4), rowspan=4,colspan=2)
# Using Numpy to create an array x
x = np.arange(1, 1001)

# Plot for exponent
plot1.plot(x, df2['Bandwidth'],'.', markeredgewidth=0.02)
plot1.set_title('Bandwidth')
idx = np.round(np.linspace(0, len(df2['Bandwidth']) - 1, 10)).astype(int)
plot1.set(yticks=df2['Bandwidth'][idx])
plt.xlabel("NÂº of interations for 2 x 2",fontsize=8)

# Plot for square root
plot4.plot(x, df2['Relative error'],'.', markeredgewidth=0.02)
plot4.set_yscale('log')
plot4.set_title('Relative error')



# Plot for Square
plot3.plot(x, df2['Compute rate (GFLOPS)'],'.', markeredgewidth=0.02)
plot3.set_title('Compute rate (GFLOPS)')
idx = np.round(np.linspace(0, len(df2['Compute rate (GFLOPS)']) - 1, 10)).astype(int)
plot3.set(yticks=df2['Compute rate (GFLOPS)'][idx])

# Plot for Square
plot2.plot(x, df2['Runtime'],'.', markeredgewidth=0.02)
plot2.set_title('Runtime (s)')
idx = np.round(np.linspace(0, len(df2['Relative error']) - 1, 10)).astype(int)
plot2.set(yticks=df2['Runtime'][idx])





# Packing all the plots and displaying them
plt.tight_layout()
plt.savefig('2.png', dpi=300, bbox_inches='tight')
```

```python
df2_NaN = df2.replace(['1.#INF00','-1.#INF00'],np.nan)
df2_NaN = df2_NaN.dropna()
df2_NaN
```
